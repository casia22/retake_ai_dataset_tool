```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_path = 'data/roth-1679831025-978929000/1679831873-527227000.jpg'
img = cv2.imread(img_path)
img = cv2.resize(img,(1920,1080))
# plot the image
plt.imshow(img),img.shape
```

```python
# center crop
height, width, _ = img.shape
crop_size = (500, 800)         

center_x = width // 2
center_y = height // 2
crop_x = center_x - crop_size[1] // 2
crop_y = center_y - crop_size[0] // 2
crop_x2 = crop_x + crop_size[1]
crop_y2 = crop_y + crop_size[0]
img = img[crop_y:crop_y2,crop_x:crop_x2,:]
img = cv2.resize(img,(200,125))
plt.imshow(img),img.shape
```
